```python
#boxplot
q1 = df['height'].quantile(.25)
q3 = df['height'].quantile(.75)
```

```python
iqr = q3 - q1
iqr
```

```python
low_whisker = q1 - 1.5 * iqr
low_whisker
```

```python
up_whisker = q3 + 1.5 * iqr
up_whisker
```

```python
def outlier(df, col):
    q1 = df['height'].quantile(.25)
    q3 = df['height'].quantile(.75)

    lower_whisker = q1 - 1.5 * iqr
    upper_whisker = q3 + 1.5 * iqr
    
    return df[(df[col] <= upper_whisker) & (df[col] >= lower_whisker)]
```

```python
df4 = outlier(df, "height")
```

```python
df4.shape
```

```python
sns.boxplot(df4['height'])
```

```python
df4 = outlier(df, "weight")
```
